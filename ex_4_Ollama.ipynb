```python
import ollama
```

```python
data = {
    "name": "Giovanni",
    "age": 30,
    "city": "Roma",
    "profession": "Ingegnere"
}

instructions = """
Trasforma i dati strutturati forniti in una descrizione testuale
completa, naturale e scorrevole.

Regole:
- utilizza tutte le informazioni presenti;
- non modificare i dati;
- non inventare informazioni aggiuntive;
- evita di mostrare nomi delle chiavi, codice Python o elenchi;
- scrivi un unico paragrafo in italiano;
- restituisci esclusivamente la descrizione.
"""

response = ollama.chat(
    model="llama3.2",
    messages=[
        {
            "role": "system",
            "content": instructions
        },
        {
            "role": "user",
            "content": str(data)
        }
    ]
)

description = response["message"]["content"].strip()

print("Descrizione:")
print(description)
```
