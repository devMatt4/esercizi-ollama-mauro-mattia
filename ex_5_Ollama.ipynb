```python
import ollama
```

```python
input_text = input(
    "Inserisci il testo su cui generare le domande:\n"
)

if not input_text.strip():
    raise ValueError("Il testo non può essere vuoto.")

instructions = """
Sei un assistente educativo specializzato nella comprensione del testo.

Leggi attentamente il testo fornito e genera 5 domande
di comprensione basate esclusivamente sul suo contenuto.

Regole:
- crea domande chiare e pertinenti;
- includi domande sui concetti e sulle informazioni principali;
- utilizza un livello di difficoltà medio;
- non inventare informazioni assenti nel testo;
- non fornire le risposte;
- numera le domande da 1 a 5;
- scrivi le domande nella stessa lingua del testo.

Restituisci esclusivamente l'elenco delle domande.
"""

response = ollama.chat(
    model="llama3.2",
    messages=[
        {
            "role": "system",
            "content": instructions
        },
        {
            "role": "user",
            "content": input_text
        }
    ]
)

questions = response["message"]["content"].strip()

print("\nDomande di comprensione:")
print(questions)
```
