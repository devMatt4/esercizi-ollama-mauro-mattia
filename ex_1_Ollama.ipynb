```python
import ollama

print("Ollama inizializzato correttamente.")
```

```python
input_text = input(
    "Inserisci una breve descrizione di prodotto in italiano o inglese:\n"
)

if not input_text.strip():
    raise ValueError("Il testo da tradurre non può essere vuoto.")

instructions = """
Sei un traduttore professionale specializzato nelle descrizioni di prodotto.

Determina automaticamente la lingua del testo fornito:
- se il testo è in italiano, traducilo in inglese;
- se il testo è in inglese, traducilo in italiano.

La traduzione deve essere:
- naturale e scorrevole;
- fedele al significato originale;
- adatta alla descrizione commerciale di un prodotto;
- priva di spiegazioni, commenti o testo aggiuntivo.

Restituisci esclusivamente la traduzione.
"""

response = ollama.chat(
    model="llama3.2",
    messages=[
        {
            "role": "system",
            "content": instructions
        },
        {
            "role": "user",
            "content": input_text
        }
    ]
)

translation = response["message"]["content"].strip()

print("\nTraduzione:")
print(translation)
```
