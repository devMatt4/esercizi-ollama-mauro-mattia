```python
import ollama
```

```python
input_text = input(
    "Inserisci un testo descrittivo relativo a una persona:\n"
)

if not input_text.strip():
    raise ValueError("Il testo non può essere vuoto.")

instructions = """
Analizza il testo fornito ed estrai le seguenti informazioni:

- nome
- età
- professione

Restituisci i dati esclusivamente nel seguente formato:

Nome: valore
Età: valore
Professione: valore

Non aggiungere spiegazioni o commenti.

Se un'informazione non è presente nel testo,
scrivi "Non specificata".
"""

response = ollama.chat(
    model="llama3.2",
    messages=[
        {
            "role": "system",
            "content": instructions
        },
        {
            "role": "user",
            "content": input_text
        }
    ]
)

print("\nInformazioni estratte:")
print(response["message"]["content"].strip())
```
