```python
import ollama
```

```python
input_text = input(
    "Inserisci l'articolo da riassumere:\n"
)

if not input_text.strip():
    raise ValueError("L'articolo non può essere vuoto.")

instructions = """
Sei un assistente specializzato nella sintesi di articoli di notizie.

Riassumi l'articolo fornito rispettando rigorosamente queste regole:

- conserva esclusivamente le informazioni più importanti;
- elimina dettagli secondari, ripetizioni e informazioni superflue;
- non aggiungere informazioni assenti nell'articolo;
- scrivi il riassunto in un unico paragrafo;
- utilizza un linguaggio chiaro, neutrale e scorrevole;
- il riassunto non deve superare i 255 caratteri, spazi inclusi;
- non inserire titoli, elenchi, introduzioni o commenti.

Restituisci esclusivamente il riassunto.
"""

response = ollama.chat(
    model="llama3.2",
    messages=[
        {
            "role": "system",
            "content": instructions
        },
        {
            "role": "user",
            "content": input_text
        }
    ]
)

summary = response["message"]["content"].strip()

# Controllo aggiuntivo per garantire il limite
if len(summary) > 255:
    summary = summary[:252].rstrip() + "..."

print("\nRiassunto:")
print(summary)

print(f"\nNumero di caratteri: {len(summary)}/255")
```
